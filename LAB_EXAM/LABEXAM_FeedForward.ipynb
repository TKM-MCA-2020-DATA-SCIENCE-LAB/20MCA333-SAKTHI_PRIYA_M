FEED FORWARD NEURAL NETWORK

```python
import keras 
import tensorflow
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

```python
(xtrain,ytrain),(xtest,ytest)=mnist.load_data()
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtest[0]
```

```python
xtrain=xtrain.reshape(60000,784).astype('float32')
xtest=xtest.reshape(10000,784).astype('float32')
```

```python
xtrain/=255
xtest/=255
```

```python
xtest[0]
```

```python
from keras import utils as np_utils
nclasses=10
```

```python
ytrain=keras.utils.np_utils.to_categorical(ytrain,nclasses)
ytest=keras.utils.np_utils.to_categorical(ytest,nclasses)
```

```python
model=Sequential()
```

```python
model.add(Dense(64,activation='sigmoid',input_shape=(784,)))
```

```python
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='MeanSquaredError',optimizer=SGD(learning_rate=0.01),metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,batch_size=40,epochs=50,verbose=1)
```

```python
ypred=model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import r2_score
print(r2_score(ytest,ypred))
```

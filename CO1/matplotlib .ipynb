```python
import numpy as np
x=np.linspace(0,5,11)
y=x ** 2 
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'r')
plt.xlabel('X axis Title Here')
plt.ylabel('Y axis Title here')
plt.title('String Title more')
plt.show()
```

```python
#plt.subplot(nrows,ncols,plot_number)
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,2)
plt.plot(x,y,'b*-')
plt.subplot(1,3,3)
plt.plot(x,y,'g--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'b')
axes.set_xlabel('Set x label')
axes.set_ylabel('Set y label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.3,0.5,0.2,0.2])

axes1.plot(x,y,'b')
axes1.set_xlabel('Axes1 x label')
axes1.set_ylabel('Axes1 y label')
axes1.set_title('Axes1 Title')

axes2.plot(x,y,'r')
axes2.set_xlabel('Axes2 x label')
axes2.set_ylabel('Axes2 y label')
axes2.set_title('Axes2 Title')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title');
```

```python
from bs4 import BeautifulSoup #python library for pulling data out of webpage
import csv

import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage =requests.get("https://www.flipkart.com/tablets/~apple-ipads/pr?sid=tyy,hry&otracker=nmenu_sub_Electronics_0_Apple%20iPads")
```

```python
webpage
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
```

```python
print(soup.prettify()) #prettify in beautiful soup will enable us to view how the tags are nested in the document
```

```python
names=soup.find_all("div",class_="_4rR01T")
```

```python
names
```

```python
ipad_names=[]
for i in range(0,len(names)):
  ipad_names.append(names[i].get_text())
```

```python
ipad_names
```

```python
ratings=soup.find_all("div",class_="_3LWZlK")
```

```python
ratings
```

```python
len(ratings)
```

```python
ipad_ratings=[]
for i in range(0,len(names)):
  ipad_ratings.append(ratings[i].get_text())
```

```python
ipad_ratings
```

```python
price=soup.find_all("div",class_="_30jeq3 _1_WHN1")
```

```python
price
```

```python
ipad_price=[]
for i in range(0,len(price)):
  ipad_price.append(price[i].get_text())
```
